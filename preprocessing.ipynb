```python
import pandas as pd
import os
```

```python
# Initialize an empty list to hold individual dataframes
dataframes = []

# Path to the folder containing CSV files
folder_path = 'data/raw'

# Loop through each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):  # Process only CSV files
        # Extract the ticker from the file name (between the parentheses)
        ticker = file_name[file_name.find('(')+1:file_name.find(')')]
        
        # Read the CSV file
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        
        # Add the ticker column
        df['ticker'] = ticker
        
        # Append the dataframe to the list
        dataframes.append(df)

# Concatenate all dataframes into a single dataframe, aligning columns and filling missing values with NaN
final_df = pd.concat(dataframes, ignore_index=True)
```

```python
final_df
```

```python
nan_percentage = (
    final_df.groupby('ticker')
    .apply(lambda group: group.isna().mean() * 100)
)

# Display the result
nan_percentage
```

```python
# Calculate the mean percentage of NaN values across all tickers for each column
overall_nan_percentage = nan_percentage.mean()

# Display the result
overall_nan_percentage
```

```python
# Identify columns with more than 40% missing data
cols_to_keep = overall_nan_percentage[overall_nan_percentage <= 40].index

# Filter the dataframe to keep only the selected columns
filtered_df = final_df[cols_to_keep]
filtered_df.dropna(inplace=True)
```

```python
# Filter tickers with at least 16 observations
ticker_counts = filtered_df['ticker'].value_counts()
tickers_to_keep = ticker_counts[ticker_counts >= 16].index

# Filter the dataframe to include only rows with tickers in the 'tickers_to_keep' list
filtered_df = filtered_df[filtered_df['ticker'].isin(tickers_to_keep)]
```

```python
filtered_df.to_csv('data/clean/filtered_data.csv', index=False)
```
